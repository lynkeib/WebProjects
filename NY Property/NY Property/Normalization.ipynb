```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data/Cleaned Data.csv', index_col = 0)
```

```python
list(df.columns)
```

```python
df.head()
```

```python
ZIP5 = df['ZIP'].astype('str').str.slice(start = 0, stop = 5)
ZIP3 = ZIP5.str.slice(stop = 3)
```

```python
del df['ZIP']
df['ZIP5'] = ZIP5
df['ZIP3'] = ZIP3
```

```python
df.head()
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df_new = df[['BBLE',
             'ZIP5',
             'ZIP3',
             'B',
             'TAXCLASS',
             'BLOCK',
             'BLDGCL',
             'FULLVAL',
             'AVLAND',
             'AVTOT',
             'LTFRONT',
             'LTDEPTH',
             'BLDFRONT',
             'BLDDEPTH',
             'STORIES']]
```

```python
df_new['ALL'] = 1
df_return = pd.DataFrame()
```

```python
df_new.head()
```

```python
variables = list(df_new.columns)[7:-1]
groups = list(df_new.columns)[1:5]
groups.append('ALL')
```

```python
variables
```

```python
for group in groups:
    for variable in variables:
        df_return[variable + '_By_' + group] = (df_new[variable] / df_new.groupby(group)[variable].transform('mean')) 
```

```python
df_return.shape
```

```python
df_return["BBLE"] = df_new['BBLE']
```

```python
col = ["BBLE"] + list(df_return.columns)[:-1]
```

```python
df_return = df_return[col]
```

```python
df_return.to_csv('~/Normalized_Data_45.csv')
```
