```python
mat = df_new.values
```

```python
pca=PCA(n_components=8)
```

```python
newData=pca.fit_transform(mat)
```

```python
newData.shape
```

```python
sum(pca.explained_variance_ratio_)
```

```python
df_pca = pd.DataFrame(newData)
```

```python
df_test = df_pca.iloc[:,:-1]
```
