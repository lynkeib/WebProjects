```python
import pandas as pd
from sklearn.decomposition import PCA
from scipy import stats
import numpy as np
from sklearn.preprocessing import scale
```

```python
df = pd.read_csv('~/D/2019_Spring/DSO562_Fraud Analytics/ASSIGNMENTS/Fraud/Normalized_Data_45.csv', index_col = 0)
```

```python
df.head()
```

```python
BBLE = df["BBLE"]
```

```python
del df['BBLE']
```

```python
df = pd.DataFrame(scale(df.values))
```

```python
columns = list(df.columns)
```

```python
df_new = df[columns]
```

```python
pca=PCA(n_components=8)
```
